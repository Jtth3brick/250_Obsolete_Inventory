```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
from typing import Dict, List

from projectCode.config import RevisionConfig, SimulationConfig
from projectCode.policies import SimpleFeasiblePolicy, SimpleSSPolicy, MyopicPolicy, OptimalPolicy
from projectCode.simulation import RevisionSimulator, SimulationResults
from projectCode.plotting import (print_performance_statistics, plot_relative_performance, 
                                plot_cost_breakdown, plot_process_analysis, plot_combined_analysis)

# Constants for experiment
TOTAL_TARGET = 80
NUM_REPLICATIONS = 5
PLANNING_HORIZON = 120

# Create configurations
rev_config = RevisionConfig(
    planning_horizon=PLANNING_HORIZON,
    total_target=TOTAL_TARGET,
    setup_cost=5000000000,
    unit_cost=100,
    holding_cost=10,
    backorder_cost=5000000000,
    lead_times=[3, 3, 3, 3, 3]
)

sim_config = SimulationConfig(
    num_replications=NUM_REPLICATIONS,
    random_seed=42
)

# Initialize simulator
simulator = RevisionSimulator(rev_config)

# Initialize policies
feasible_policy = SimpleFeasiblePolicy(rev_config)
ss_policy = SimpleSSPolicy(reorder_point=3, order_up_to=20, config=rev_config)
myo_policy = MyopicPolicy(rev_config)
optimal_policy = OptimalPolicy(config=rev_config)

# Run simulations
policies = {
    "SimpleFeasible": feasible_policy,
    "SimpleSSPolicy": ss_policy,
    "MyopicPolicy": myo_policy,
    "OptimalPolicy": optimal_policy
}

results = {}
for name, policy in policies.items():
    print(f"\nEvaluating {name}...")
    policy_results = simulator.evaluate_policy(
        policy,
        num_episodes=NUM_REPLICATIONS,
        seeds=list(range(NUM_REPLICATIONS))
    )
    results[name] = policy_results

# Print detailed statistics
stats = print_performance_statistics(results)

# Plot relative performance and save
fig1 = plot_relative_performance(results)
plt.savefig('analysis_output/relative_performance.png', bbox_inches='tight', dpi=300)

# Plot cost breakdown and save
fig2 = plot_cost_breakdown(results)
plt.savefig('analysis_output/policy_breakdown.png', bbox_inches='tight', dpi=300)

# Plot cumulative analysis and save
# Note: plot_cumulative_analysis expects single results, not lists
single_results = {name: results[name][0] for name in results}
fig3 = plot_process_analysis(single_results)
plt.savefig('analysis_output/cumulative_analysis.png', bbox_inches='tight', dpi=300)

plt.show()
```
